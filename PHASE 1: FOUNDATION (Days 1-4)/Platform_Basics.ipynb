```python
# Creating DataFrame
data = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(data, ["product", "price"])
df.show()
```

```python
# Filtering expensive products
df.filter(df.price > 1000).show()
```

```python
# Sorting by price
df.orderBy(df.price.desc()).show()
```

```python
# Adding a new column
from pyspark.sql.functions import when

df2 = df.withColumn(
    "category",
    when(df.price > 1000, "Premium").otherwise("Standard")
)

df2.show()
```

```python
# Perforing Basic aggregation
df2.groupBy("category").count().show()
```
